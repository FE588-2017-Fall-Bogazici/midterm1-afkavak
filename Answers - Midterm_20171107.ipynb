```python
## Q1

# function to estimate moving average of list lst, window L

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

def movingAverage(lst,L):
    seqMA=[]
    for i in range(len(lst)):
        if(i-L+1<0):
            j=0
        else:
            j=i-L+1
        seqMA.append(sum(lst[j:i+1])/L)
    return(seqMA)

# lst1=[1,4,5,-1]
# movingAverage(lst1,3)

# function to generate a random walk with length L
def generateRW(L):
    seriesRW=[]
    for i in range(L):
        seriesRW.append(np.random.randn())
    return(seriesRW)
plt.plot(movingAverage(generateRW(200),200))
```
